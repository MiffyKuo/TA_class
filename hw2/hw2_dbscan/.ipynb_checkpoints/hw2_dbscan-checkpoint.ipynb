```python
import pandas as pd
import numpy as np
import math

UNCLASSIFIED = False
NOISE = -1
```

```python
def _dist(p,q):
    dist = np.linalg.norm(p-q)
    return dist
```

```python
def _eps_neighborhood(p,q,eps):
    return _dist(p,q) < eps
```

```python
def _region_query(m, point_id, eps):
    n_points = m.shape[1]
    seeds = []
    for i in range(0, n_points):
        if _eps_neighborhood(m[:,point_id], m[:,i], eps):
            seeds.append(i)
    return seeds                 #圈圈裡的樣本數的 id(第幾個點)
```

```python
def _expand_cluster(m, classifications, point_id, cluster_id, eps, min_points):
    # expand the cluster of cluster_id from point_id
    # identify all points belonging to cluster_id
    # update the clustering results in "classification" by assigning cluster_id to each point
    # return True if point_id is a core, False if point_id is not a core
    # write your code below
    
    seeds = _region_query(m, point_id, eps)
    if len(seeds) < min_points:             #圈圈裡的樣本數小於 min points, 就將那個點的 point_id 改成 NOISE(=-1)
        classifications[point_id] = NOISE
        return False
    else:                                   #圈圈裡的樣本數大於等於 min points
        classifications[point_id] = cluster_id    #將目標點的 id改成 cluster id                           
        for seed_id in seeds:                     
            classifications[seed_id] = cluster_id     #將 seeds裡點的 id都先改成和目標點一樣(有問題在 while loop會改)
            
        while len(seeds) > 0:
            current_point = seeds[0]                        #用 seeds的第一個點當目標點
            results = _region_query(m, current_point, eps)  #繼續用 seeds裡第一個點畫圈圈
            if len(results) >= min_points:             #如果畫出來的圈圈裡的樣本數 >= min points
                for i in range(0, len(results)):            #用一個 for loop跑完所有圈圈裡的樣本
                    result_point = results[i]                                    
                    if classifications[result_point] == UNCLASSIFIED or \
                       classifications[result_point] == NOISE:
                        if classifications[result_point] == UNCLASSIFIED:       #如果這個點的 id為 Unclassified
                            seeds.append(result_point)                          #就將這個點 append到這個目標點的 seeds裡
                        classifications[result_point] = cluster_id          #然後將這個點的 id改成 cluster id(分群)
            seeds = seeds[1:]                       # seeds裡第一個點跑完後把第一個點去掉，繼續判斷第二個，直到所有 seeds判斷完                                                                   
        return True
    
    
    
    
```

```python
def dbscan(m, eps, min_points):
    """Implementation of DBSCAN
    You can refer to wikipedia for detailed algorithm: https://en.wikipedia.org/wiki/DBSCAN
    Use Euclidean Distance as the measure
    
    Inputs:
    m - A matrix whose columns are feature vectors
    eps - Maximum distance two points can be to be regionally related
    min_points - The minimum number of points to make a cluster
    
    Outputs:
    An array with either a cluster id number or dbscan.NOISE (None) for each column vector in m
    """
    cluster_id = 1                                #第一群 (有新的群後面會改變 cluster id )
    n_points = m.shape[1]                         #有幾個點(80)
    classifications = [UNCLASSIFIED] * n_points   #將 classifications設為一個 list，第一隔代表第一個點的分群，先將所有點都設為 Unclassified
    # the main dbscan algorithm
    # put your code here
    
    for point_id in range(0, n_points):
        point = m[:,point_id]                     #用 for loop跑完 m找出每個點
        if classifications[point_id] == UNCLASSIFIED:                                     #如果這個點還沒被分群
            if _expand_cluster(m, classifications, point_id, cluster_id, eps, min_points):    #跑一次 expand cluster func，如果回傳是 True
                cluster_id = cluster_id + 1                                                       #就將 cluster id + 1 (第 2 群等...)
    return classifications                                                                    #最後回傳每一個點的 classification (畫圖用)

    
    return classifications
```

```python
# test here
dataset_1 = pd.read_csv('blobs.csv')[:80].values
m = np.asmatrix(dataset_1)
m = m.transpose()
m[:,0]
```

```python
eps = 1.6
min_points = 5
a = dbscan(m, eps, min_points)
```
